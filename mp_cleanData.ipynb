```python
import pandas as pd

df= pd.read_csv("datasets/urad.csv", skiprows=1) 
```

```python
df.head(10)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.rename(columns={
    "Arrival Quantity 01-01-2021 to 31-03-2026": "Arrival",
    "Modal Price 01-01-2021 to 31-03-2026": "Price"
}, inplace=True)
```

```python
df.head()
```

```python
commodity_mapping = {
    "Paddy(Common)": "Rice",
    "Maize": "Maize",
    "Jowar(Sorghum)": "Jowar",
    "Bajra(Pearl Millet/Cumbu)": "Bajra",
    "Groundnut": "Groundnut",
    "Black Gram(Urd Beans)(Whole)": "Urad",
    "Green Gram(Moong)(Whole)": "Moong"
}
```

```python
df["Commodity"] = df["Commodity"].replace(commodity_mapping)
```

```python
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors='coerce')
df["Arrival"] = pd.to_numeric(df["Arrival"], errors='coerce')
df["Price"] = pd.to_numeric(df["Price"], errors='coerce')
```

```python
df.head()
```

```python
df = df.dropna()
df = df[(df["Arrival"] > 0) & (df["Price"] > 0)]
```

```python
df.shape[0]
```

```python
df.to_csv("urad1.csv", index=False)
```

```python
import pandas as pd
import os

DATA_FOLDER = "."   # current directory

all_dfs = []

for file in os.listdir(DATA_FOLDER):
    if file.endswith(".csv"):   # better than "1.csv"
        path = os.path.join(DATA_FOLDER, file)
        df = pd.read_csv(path)
        all_dfs.append(df)

# Merge all
global_df = pd.concat(all_dfs, ignore_index=True)

# Save
global_df.to_csv("global_dataset.csv", index=False)

print("Merged all crops!")
print("Total rows:", global_df.shape)
```
